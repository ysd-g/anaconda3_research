```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina' #グラフを高解像度で表示

import matplotlib.pyplot as plt
import pandas as pd #ファイルからデータを読み込む際に必要
from sklearn import linear_model #sklearnは、機械学習のライブラリ
```

```python
dataframe = pd.read_fwf('brains_body.txt')
x_values = dataframe[['Brain']]
y_values = dataframe[['Body']]
# brains_body.txtの「Brain」列の2行目以降の値をx_valuesに入れる
# brains_body.txtの「Body」列の2行目以降の値をy_valuesに入れる
```

```python
#x_values（脳の重量）から、y_values（体重）を推定する線形モデル
body_reg = linear_model.LinearRegression()
body_reg.fit(x_values, y_values)#最もうまく近似する一次関数を探す
```

```python
plt.scatter(x_values, y_values) #散布図
plt.plot(x_values, body_reg.predict(x_values)) #直線を描く
plt.show() #グラフを表示
```
